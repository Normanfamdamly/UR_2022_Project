```python
# Read MentalHealth_StateData.csv 
clean_mh1_data = "MH1_Data.csv"
clean_mh1_df = pd.read_csv(clean_mh1_data)
```

```python
# Convert columns STATEIP, DIVISION and REGION from int64 to string
clean_mh1_data_df['MH1']=clean_mh1_data_df['MH1'].map(str)
```

```python
# Convert MH1 from numbers to diagnosis
clean_mh1_data_df['MH1']=clean_mh1_data_df['MH1'].replace(['1'], 'Trauma and Stress')
clean_mh1_data_df['MH1']=clean_mh1_data_df['MH1'].replace(['2'], 'Anxiety')
clean_mh1_data_df['MH1']=clean_mh1_data_df['MH1'].replace(['3'], 'ADHD')
clean_mh1_data_df['MH1']=clean_mh1_data_df['MH1'].replace(['4'], 'Conduct')
clean_mh1_data_df['MH1']=clean_mh1_data_df['MH1'].replace(['5'], 'Delirium, Dementia')
clean_mh1_data_df['MH1']=clean_mh1_data_df['MH1'].replace(['6'], 'Bipolar')
clean_mh1_data_df['MH1']=clean_mh1_data_df['MH1'].replace(['7'], 'Depressive')
clean_mh1_data_df['MH1']=clean_mh1_data_df['MH1'].replace(['8'], 'Oppositional Defiant ')
clean_mh1_data_df['MH1']=clean_mh1_data_df['MH1'].replace(['9'], 'Pevassive Developmental')
clean_mh1_data_df['MH1']=clean_mh1_data_df['MH1'].replace(['10'], 'Personality')
clean_mh1_data_df['MH1']=clean_mh1_data_df['MH1'].replace(['11'], 'Schizophrenia')
clean_mh1_data_df['MH1']=clean_mh1_data_df['MH1'].replace(['12'], 'Alcohol or Substance')
clean_mh1_data_df['MH1']=clean_mh1_data_df['MH1'].replace(['13'], 'Other')
clean_mh1_data_df['MH1']=clean_mh1_data_df['MH1'].replace(['-9'], 'Unknown or Deferred')
```

```python
# Save DataFrame as a new file
clean_mh1_data_df.to_csv('MH1_Data.csv')
```
