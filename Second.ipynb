```python
import pandas as pd
import numpy as np
```

```python
# Read mhcld-puf-2020-csv.csv and create DataFrame 
mental_health_data = "Resources/mhcld-puf-2020-csv.csv"
mh_df = pd.read_csv(mental_health_data)
mh_df.head()
```

```python
# List of all the columns
mh_df.columns
```

```python
# Clean up the data
clean_mh_df = mh_df.drop(columns=['EDUC', 'RACE', 'SPHSERVICE',
       'CMPSERVICE', 'OPISERVICE', 'RTCSERVICE', 'IJSSERVICE', 'MH1', 'MH2',
       'MH3', 'SUB', 'MARSTAT', 'SMISED', 'SAP', 'EMPLOY', 'DETNLF', 'VETERAN',
       'LIVARAG', 'NUMMHS', 'TRAUSTREFLG', 'ANXIETYFLG', 'ADHDFLG',
       'CONDUCTFLG', 'DELIRDEMFLG', 'BIPOLARFLG', 'DEPRESSFLG', 'ODDFLG',
       'PDDFLG', 'PERSONFLG', 'SCHIZOFLG', 'ALCSUBFLG', 'OTHERDISFLG'])
```

```python
clean_mh_df['AGE']=clean_mh_df['AGE'].map(str)
```

```python
# Replace the age from the dataframe to match the actual data from the code book
clean_mh_df['AGE']=clean_mh_df['AGE'].replace(['1'], '5')
clean_mh_df['AGE']=clean_mh_df['AGE'].replace(['2'], '13')
clean_mh_df['AGE']=clean_mh_df['AGE'].replace(['3'], '16')
clean_mh_df['AGE']=clean_mh_df['AGE'].replace(['4'], '19')
clean_mh_df['AGE']=clean_mh_df['AGE'].replace(['5'], '22')
clean_mh_df['AGE']=clean_mh_df['AGE'].replace(['6'], '27')
clean_mh_df['AGE']=clean_mh_df['AGE'].replace(['7'], '32')
clean_mh_df['AGE']=clean_mh_df['AGE'].replace(['8'], '37')
clean_mh_df['AGE']=clean_mh_df['AGE'].replace(['9'], '42')
clean_mh_df['AGE']=clean_mh_df['AGE'].replace(['10'], '47')
clean_mh_df['AGE']=clean_mh_df['AGE'].replace(['11'], '52')
clean_mh_df['AGE']=clean_mh_df['AGE'].replace(['12'], '57')
clean_mh_df['AGE']=clean_mh_df['AGE'].replace(['13'], '62')
clean_mh_df['AGE']=clean_mh_df['AGE'].replace(['14'], '65')
clean_mh_df['AGE']=clean_mh_df['AGE'].replace(['-9'], 'NaN')
clean_mh_df.head()
```
